```python
#Reading File
df=pd.read_csv(r"C:\Users\ASUS\Desktop\Yasmin AI Training\cleaning project\Orders.csv")
df
```

```python
df.info() #dataFrame info
```

```python
df.dropna(how='all',inplace=True) # dropping NAN Values
df
```

```python
#convert city name To upper case
df['City']=df['City'].str.title()
df
```

```python
#delet 'Tel' string from phone number column
df['Phone Number']=df['Phone Number'].str.replace('Tel:','')
df
```

```python
# rounding decimal value 
df['Discount']=df['Discount'].round(0)
df
```

```python
#Infer new columns from data
  #1- infer 'total price' column
df['Total Price']=df['Quantity']*df['Unit Price']
  #2- infer 'Cost' column
df['Cost']=df['Total Price']-df['Discount']
df
```
